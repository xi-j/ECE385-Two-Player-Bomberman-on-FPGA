# Audio to Hex in Flash

```python
from IPython.display import Audio
from scipy.io import wavfile
import numpy as np

def tohex(val, nbits):
  return hex((val + (1 << nbits)) % (1 << nbits))
```

```python
fs, audio_Back = wavfile.read('Back_ground.wav')
fs, audio_Bomb = wavfile.read('Bomb_Sound.wav')

audio_Back_L = audio_Back[:,0]
audio_Back_R = audio_Back[:,1]

audio_Bomb = audio_Bomb[:,0]


Audio(data = audio_Back_L, rate = fs)


```

```python
audio_out_name = "./audio_out/audio_whole.hex" 

audio_out = open(audio_out_name,'w')
audio_list_L = []
audio_list_R = []


if (len(audio_Bomb) > (8388607-7340032)):
    print("Bomb too big!")
    assert(1)


for r in range (0,len(audio_Back_L)):
    
    hex_l = tohex(audio_Back_L[r],16) 
    hex_l = hex_l[2:].zfill(4)
    hex_r = tohex(audio_Back_R[r],16) 
    hex_r = hex_r[2:].zfill(4)
    
    audio_out.write(hex_l[0:2])
    audio_out.write(hex_l[2:])
#     audio_out.write("\n")
    audio_out.write(hex_r[0:2])
    audio_out.write(hex_r[2:])
#     audio_out.write("\n")

    
    audio_list_L.append(hex_l)
    audio_list_R.append(hex_r)

    
for z in range (len(audio_Back_L)*2,3670016):
    audio_out.write("0000")
#     audio_out.write("\n")

for r in range (3670016,3670016+len(audio_Bomb)):
    
    hex_l = tohex(audio_Bomb[r-3670016],16) 
    hex_l = hex_l[2:].zfill(4)
    
    audio_out.write(hex_l[0:2])
    audio_out.write(hex_l[2:])


for k in range (3670016+len(audio_Bomb),4194304):
    audio_out.write("0000")
#     audio_out.write("\n")


print("Finish")

audio_out.close()
```

```python
audio_in_name = "./audio_out/audio_whole.hex" 
audio_out_name0 = "./audio_out/audio_whole0.hex" 
audio_out_name1 = "./audio_out/audio_whole1.hex" 
audio_out_name2 = "./audio_out/audio_whole2.hex" 
audio_out_name3 = "./audio_out/audio_whole3.hex" 
audio_out_name4 = "./audio_out/audio_whole4.hex" 

audio_in = open(audio_in_name,'r')
audio_out0 = open(audio_out_name0,'w')
audio_out1 = open(audio_out_name1,'w')
audio_out2 = open(audio_out_name2,'w')
audio_out3 = open(audio_out_name3,'w')
audio_out4 = open(audio_out_name4,'w')

for z in range(0,2097152*2):
    file_in = audio_in.read(1)
    audio_out0.write(file_in)
    
for z in range(2097152*2,2097152*4):
    file_in = audio_in.read(1)
    audio_out1.write(file_in)
    
for z in range(2097152*4,2097152*6):
    file_in = audio_in.read(1)
    audio_out2.write(file_in)
    
for z in range(2097152*6,2097152*7):
    file_in = audio_in.read(1)
    audio_out3.write(file_in)
    
for z in range(2097152*7,2097152*8):
    file_in = audio_in.read(1)
    audio_out4.write(file_in)
    

audio_in.close()
audio_out0.close()
audio_out1.close()
audio_out2.close()
audio_out3.close()
audio_out4.close()
print("Finish")
```
